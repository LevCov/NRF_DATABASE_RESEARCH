```python
import pandas as pd
```

```python
m_bench = pd.read_csv("/home/anx_tl/bench_data/mongo_bench.csv")
r_bench = pd.read_csv("/home/anx_tl/bench_data/redis_bench.csv")
rocks_bench = pd.read_csv("/home/anx_tl/bench_data/rocks_bench.csv")
mysql_bench = pd.read_csv("/home/anx_tl/bench_data/mysql_bench.csv")
crud_couch = pd.read_csv("/home/anx_tl/bench_data/bench_crud.csv")
search_couch = pd.read_csv("/home/anx_tl/bench_data/bench_serch.csv")
```

```python
m_create = m_bench[m_bench['name'].str.contains(r'^(BM_Create)')][:5]
r_create = r_bench[r_bench['name'].str.contains(r'^(BM_Create)')][:5]
rocks_create = rocks_bench[rocks_bench['name'].str.contains(r'^(BM_Create)')][:5]
mysql_create = mysql_bench[mysql_bench['name'].str.contains(r'^(BM_Create)')][:5]
couch_create = crud_couch[crud_couch['name'].str.contains(r'^(BM_Create)')][:5]

create = pd.DataFrame(columns=['mongo', 'redis', 'rocks', 'mysql', 'couch'])
create['mongo'] = m_create['real_time']
create['redis'] = r_create['real_time']
create['rocks'] = rocks_create['real_time']
create['mysql'] = mysql_create['real_time']
create['couch'] = couch_create['real_time']

x_tick = [str(i) for i in range(1, create.shape[0] + 1)]

create.plot(color=['orange', 'green', 'blue', 'red', 'yellow'], title="CREATE", ylim=(0, 170000000), xlabel='Итерация', ylabel='Время, ns').set_xticks(create.index, x_tick);
```

```python
m_find = m_bench[m_bench['name'].str.contains(r'^(BM_find)')]
m_find = m_find[m_find['name'].str.contains(r'3$')]
r_find = r_bench[r_bench['name'].str.contains(r'^(BM_find)')]
r_find = r_find[r_find['name'].str.contains(r'3$')]
rocks_find = rocks_bench[rocks_bench['name'].str.contains(r'^(BM_find)')]
rocks_find = rocks_find[rocks_find['name'].str.contains(r'3$')]
mysql_find = mysql_bench[mysql_bench['name'].str.contains(r'^(BM_find)')]
mysql_find = mysql_find[mysql_find['name'].str.contains(r'3$')]
couch_find = search_couch[search_couch['name'].str.contains(r'^(BM_Search)')]
print(couch_find)
couch_find = couch_find[couch_find['name'].str.contains(r'3$')]

idx = ['1-50', '2-50', '3-50',
       '1-100', '2-100', '3-100',
       '1-250', '2-250', '3-250',
       '1-500', '2-500', '3-500']
find = pd.DataFrame(columns=['mongo', 'redis', 'rocks', 'mysql', 'couch'])
find['mongo'] = m_find['real_time'].reset_index(drop= True)
find['redis'] = r_find['real_time'].reset_index(drop= True)
find['rocks'] = rocks_find['real_time'].reset_index(drop= True)
find['mysql'] = mysql_find['real_time'].reset_index(drop= True)
find['couch'] = couch_find['real_time'].reset_index(drop= True)
find['idx'] = idx
find.set_index('idx', inplace=True)
#x_tick = [str(i) for i in range(1, find.shape[0] + 1)]

ax = find.plot(color=['orange', 'green', 'blue', 'red', 'yellow'], title='FIND', ylim=(0, 8000000), xlabel='Итерация', ylabel='Время, ns')#.set_xticks(find.index, x_tick);
ax.set_xticks(range(len(find)))
ax.set_xticklabels(idx, rotation=45)
```
